```python
import scipy.linalg as splin
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
plt.figure(figsize = (7,7))
x = np.linspace(-1,1,100)
plt.plot(x, (1. - x**2)**.5, label = 'L2 Norm')
plt.plot(x, -(1. - x**2)**.5, color = 'b')
plt.plot(x[50:100] , (1. - x[50:100]))
plt.plot(x[50:100] , -(1. - x[50:100]), color = 'g', label = 'L1 Norm')
plt.plot(x[0:50] , 2 - (1. - x[0:50]), color = 'g')
plt.plot(x[0:50] , (1. - x[0:50]) - 2, color = 'g')
plt.xlim(-1.25, 1.25)
plt.ylim(-1.25, 1.25)
plt.legend()
plt.title("L1 and L2 Norm where k = 1")
```
