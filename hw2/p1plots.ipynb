```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline  
plt.style.use('ggplot')
```

```python
mvpdf = lambda x, mu, cov: stats.multivariate_normal.pdf(x, mu, cov)
```

```python
# multi-variate normal density
def mvdense(mu, cov):
    density = np.zeros((100,100))
    for i in range(100):
        for j in range(100):
            density[i,j] = mvpdf([i/10 -5, j/10 - 5], mu, cov)
    return density

# multi-variate plot
def mvdenseplt(mu, cov):
    plt.figure(figsize=(10,10))
    plt.imshow(mvdense(mu,cov))
    plt.title("Part a)")
    plt.xlabel("x from -10 to 10")
    plt.ylabel("y from -10 to 10")
    plt.xticks([])
    plt.yticks([])
```

```python
sigma11 = np.array([[6, 8], [8, 13]])
mu1 = np.array([0,0])

mvdenseplt(mu1, sigma11)
```

```python
mu2 = 5
sigma22 = 14

plt.figure(figsize=(10,6))

x = np.linspace(-100, 100, 200)
plt.plot(x, stats.norm.pdf(x, mu2, sigma22))
plt.ylim([-.001,.03])
plt.xlim([-80,80])
plt.title("Part b)")
```
